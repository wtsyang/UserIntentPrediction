```python
import random
```

```python
all_file = 'data/Experiment/Train1.csv'
gt = 'data/Experiment/ground_truth.csv'
```

```python
combo_dict = {}
with open(all_file, encoding="utf8") as fin:
    for line in fin:
        if line != '\n':
            tokens = line.strip().split('\t')
            labels = tokens[0]
            
            labels = labels.split('_')
            if len(labels) > 1 and 'GG' in labels:
                labels.remove('GG')
            if len(labels) > 1 and 'O' in labels:
                labels.remove('O')
            if len(labels) > 1 and 'JK' in labels:
                labels.remove('JK')
            
            labels = '_'.join(sorted(labels))
            
            combo_dict[labels] = combo_dict.setdefault(labels, 0) + 1
```

```python
sorted_combos = sorted(combo_dict.items(), key=lambda x: x[1], reverse=True)
label_dict = {item[0]: 1 for item in sorted_combos[:32]}
```

```python
label_set = set()
with open(all_file, encoding="utf8") as fin, open(gt, 'w') as fout:
    for line in fin:
        if line != '\n':
            tokens = line.strip().split(',')
            print(tokens)
            print(tokens[1])
            labels = [tokens[1],tokens[2],tokens[3],tokens[4],tokens[5],tokens[6],tokens[7],tokens[8],tokens[9],tokens[10],tokens[11],tokens[12]]
            print(labels)
            utterance = tokens[1]
            uoa = tokens[2]
            
#             labels = labels.split('_')
            if len(labels) > 1 and 'GG' in labels:
                labels.remove('GG')
            if len(labels) > 1 and 'O' in labels:
                labels.remove('O')
            if len(labels) > 1 and 'JK' in labels:
                labels.remove('JK')
            
            labels = '_'.join(sorted(labels))
            if labels not in label_dict:
                labels = labels.split('_')
                labels = random.choice(labels)
            label_set.add(labels)
            fout.write('{}\t{}\t{}\n'.format(labels, utterance, uoa))
        else:
            fout.write('\n')
```
